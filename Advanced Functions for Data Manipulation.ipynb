```python
def year(val):
    return val.split()[-1]

data['year_sold'] = data['Date House was Sold'].map(year)
data['year_sold'].head()
```

```python
def lux_home(row):
    count = 0
    if row[0] == 'Yes':
        count = count +1
    if row[1] in ['Good' , 'Excellent']:
        count = count + 1
    if row[2] >= 8:
        count = count + 1
    if count >=2 :
        return 'Yes'
    else:
        return 'No'

data['Luxury_home'] = data[['Waterfront View','Condition of the House', 'Overall Grade']].apply(lux_home,axis=1)

data['Luxury_home'].unique()
```
